```python
X = train_df.iloc[: ,0:-1 ]

x_test = test_df.iloc[: ,0:-1 ]

x_test.head()
```

```python
X = (X - X.mean())/(X.max()-X.min())
x_test = (x_test - x_test.mean())/(x_test.max()-x_test.min())
X['ones']=1
x_test['ones']=1
X
```

```python
X = X.iloc[:,0:76].values
x_test = x_test.iloc[:,0:76].values
X.shape
```

```python
y = train_df.iloc[:,len(train_df.columns)-1:len(train_df.columns)].values
theta = np.zeros([1,len(train_df.columns)])
y2 = test_df.iloc[:,len(test_df.columns)-1:len(test_df.columns)].values
alpha = 0.0075
iters = 1500
```

```python
def computeCost(X,y,theta):
    error = np.power(((X @ theta.T)-y),2)
    return np.sum(error)/(2 * len(X))
```

```python
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost

#running the grdaient descent and cost function
g,cost = gradientDescent(X,y,theta,iters,alpha)
finalCost = computeCost(X,y,g)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  
ax.plot(np.arange(iters), cost, 'g')  
ax.set_xlabel('Iterations')  
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Iterations') 
```
