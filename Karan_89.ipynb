```python
# Importing useful libraries
import pandas as pd
import numpy as np

pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
# Importing the dataset
train_data = pd.read_csv("Trainfile.csv")
test_data = pd.read_csv("Testfile.csv")
```

```python
train_data.shape
```

```python
train_data.head()
```

```python
# Dropping Idx Column
train_data = train_data.drop(['Idx'],axis=1)
test_data = test_data.drop(['Idx'],axis=1)
```

```python
# Dropping columns with missing values 
cols_with_missing_data = [col for col in train_data.columns if train_data[col].isnull().any()] 
train_data = train_data.drop(cols_with_missing_data, axis=1)
test_data = test_data.drop(cols_with_missing_data, axis=1)
```

```python
# Getting a list of categorical variables
s = (train_data.dtypes == 'object')
object_cols = list(s[s].index)
```

```python
# Applying One Hot Encoding to the categorical variables

from sklearn.preprocessing import OneHotEncoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

OH_train_data = pd.DataFrame(OH_encoder.fit_transform(train_data[object_cols]))
OH_test_data = pd.DataFrame(OH_encoder.transform(test_data[object_cols]))

# Removing the original columns having categorical variables
train_data = train_data.drop(object_cols, axis=1)
test_data = test_data.drop(object_cols, axis=1)

# Adding dummy variable columns to the dataset
train_data = pd.concat([train_data, OH_train_data], axis=1)
test_data = pd.concat([test_data, OH_test_data], axis=1)
```

```python
train_data
```

```python
# Finding variables highly correlated with SalePrice
corr_matrix = train_data.corr()
corr_matrix = corr_matrix.sort_values('SalePrice')

# Setting 0.3 as a threshold to select the required features
cols1 = corr_matrix['SalePrice'][corr_matrix['SalePrice'].values > 0.3 ].index.values
cols2 = corr_matrix['SalePrice'][corr_matrix['SalePrice'].values < -0.3 ].index.values
features = list(cols1) + list(cols2)
features.remove('SalePrice')
```

```python
# Splitting into independent and dependent variables
X_train = train_data[features]
X_test = test_data[features]
y_train = train_data['SalePrice']
y_test = test_data['SalePrice']
```

```python
def normaliser (df_train,df_test):
    
    temp_train = df_train.copy()
    temp_test = df_test.copy()
    
    for col in df_train.columns:
    
        max_value = df_train[col].max()
        min_value = df_train[col].min()
    
        temp_train[col] = (temp_train[col] - min_value) / (max_value - min_value)
        temp_test[col] = (temp_test[col] - min_value) / (max_value - min_value)

    return temp_train , temp_test

# Performing Normalisation
normalized_X_train , normalized_X_test = normaliser(X_train,X_test)
```

```python
# Adding a column of ones for constant term

final_X_train = pd.DataFrame(np.concatenate((np.ones((normalized_X_train.shape[0],1)),normalized_X_train),axis=1))

final_X_test = pd.DataFrame(np.concatenate((np.ones((normalized_X_test.shape[0],1)),normalized_X_test),axis=1))
```

```python
final_X_train.head()
```

```python
# Finding coefficients (parameters) using the normal equation method
def normal_equation (X_train,y_train):
    return (np.linalg.inv(X_train.T@X_train))@(X_train.T)@y_train

theta = normal_equation(final_X_train,y_train) 
```

```python
y_pred = final_X_test @ theta
```

```python
# Model Evaluation

# Mean Absolute Percentage Error
def mape (y,y_pred):
    mape = ( 100 / len(y) ) * sum( abs( (y-y_pred) / y ) )
    return mape
```

```python
print("Mean absolute Percentage Error =",mape(y_test,y_pred))
```
